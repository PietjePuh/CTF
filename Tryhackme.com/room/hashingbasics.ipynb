## üåà What is a Rainbow Table?

A **Rainbow Table** is a lookup table mapping hashes to plaintext passwords. It allows you to quickly discover what password corresponds to a given hash.

A rainbow table **trades cracking time for hard disk space**‚Äîit‚Äôs fast to use but slow to generate. The idea: precompute hashes for a huge number of possible passwords, store them in a table, and then simply look up a hash when you want to crack it.

### Should you encrypt passwords in password-verification systems? Yea/Nay

<details>
Answer: Nay

#### Why not encryption?

- **Encryption is reversible.** Anyone with the encryption key can recover all passwords if the system is compromised.
- **Hashing is one-way.** Properly-hashed passwords cannot be reversed back to plaintext, even if hashes are stolen.
- **Password verification only needs to check for a match,** not retrieve the original password.  
- **Industry best-practice:** Always use a slow, salted hash function (like bcrypt, scrypt, or Argon2).

##### Example:
- **Bad:** `AES_ENCRYPT(password)` ‚Üí If key is leaked, all passwords are exposed.
- **Good:** `bcrypt(password + salt)` ‚Üí Hash is useless for attackers, even if stolen.

**Bottom line:**  
> _Never encrypt passwords for verification. Always hash (with a salt) and compare hashes. That‚Äôs why the answer is **Nay**._
</details>

### What is the hash size in yescrypt?
Question Hint: Research online or check the manual page: man 5 crypt
<details>go to https://hashcat.net/wiki/doku.php?id=example_hashes > You can see this by examining a yescrypt hash (the actual hash portion is 43 base64 characters, which encodes 32 bytes). 32 bytes (256 bits)

Hoe werkt base64?
Base64 codeert 3 bytes (24 bits) als 4 tekens.

Elk base64-teken staat dus voor 6 bits.

43 base64-tekens √ó 6 = 258 bits (iets meer dan 32 bytes, vanwege padding/afronding).

Waarom toch 32 bytes?
De originele hash is 32 bytes = 256 bits (dat is wat yescrypt oplevert).
Bij het opslaan wordt die hash base64-gecodeerd om hem als leesbare tekst te kunnen tonen.
Base64 maakt het wat langer (meer tekens), maar het originele aantal bytes verandert niet.

Rekensom:

Aantal bytes: 32

Aantal bits: 32 √ó 8 = 256 bits

Aantal base64 tekens voor 32 bytes:

32 bytes √ó (4/3) = 42,66... ‚Üí afgerond 43 tekens (soms met padding).

TL;DR
yescrypt hash size: 32 bytes = 256 bits

Base64 output: 43 tekens, want base64 maakt het langer, maar verandert niet het aantal ruwe bytes.

</details>

```python
# Generate 32 bytes, encode as base64, and compare lengths

import os
import base64

# 1. Generate a 32-byte (256-bit) value (simulate a yescrypt hash)
hash_bytes = os.urandom(32)
print("Raw bytes:", hash_bytes)
print("Number of bytes:", len(hash_bytes))

# 2. Encode to base64 for storage/display
hash_base64 = base64.b64encode(hash_bytes)
print("Base64 string:", hash_base64.decode())
print("Number of base64 characters:", len(hash_base64.decode()))
```
